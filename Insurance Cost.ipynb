```python
import pandas as pd
import seaborn as sns
import numpy as np
import statsmodels.api  as sm
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\Owner\Desktop\data\Lin Reg\insurance.csv")
```

```python
df
```

```python
X = df["bmi"]
Y = df["charges"]
```

```python
sns.regplot(x="bmi", y="charges", data=df, line_kws={'color': 'red'});
```

```python
X = sm.add_constant(X) 
results = sm.OLS(Y,X).fit()
results.summary()
```

```python
X1 = sm.add_constant(df[['bmi', 'age']])
Y1 = df['charges']
```

```python
sns.regplot(x="age", y="charges", data=df, line_kws={'color': 'red'});
```

```python
results = sm.OLS(Y1, X1).fit()
results.summary()
```
